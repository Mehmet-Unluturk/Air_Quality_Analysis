```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("openaq_data_eu_l.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
conn = sqlite3.connect('air_quality.db')
```

```python
df.to_sql('air_quality', conn, index=False, if_exists='replace')
```

```python
query1 = "SELECT * FROM air_quality WHERE country = 'Greece'"
pd.read_sql_query(query1, conn)
```

```python
query2 = "SELECT * FROM air_quality WHERE country = 'Spain' and quality_label = 'Unhealthy'"
pd.read_sql_query(query2, conn)
```

```python
query3= "SELECT country, COUNT(*) AS num_records FROM air_quality GROUP BY country"
pd.read_sql_query(query3, conn)
```

```python
query4 = "SELECT quality_label, COUNT(*) AS count FROM air_quality GROUP BY quality_label"
pd.read_sql_query(query4, conn)
```

```python
query5 = "SELECT parameter, quality_label, COUNT(*) AS sample_count FROM air_quality GROUP BY parameter, quality_label"
pd.read_sql_query(query5, conn)
```
